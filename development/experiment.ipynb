```python
import collections
import abc
import random
from copy import copy
random.seed(1)

class Card():
    card_template = collections.namedtuple("Card",["rank","suit"])

    def __init__(self, rank, suit):
        self.card = self.card_template[rank,suit]
        
    def __repr__(self):
        return f"{getattr(self.card,"rank")} of {getattr(self.card,"suit")}"

class BaseDeck(abc.ABC):
    def __init__(self):
        self.cards = self.build_deck()
        self.free_cards = self.cards

    @abc.abstractmethod
    def build_deck(self):
        """Should build a deck specific to a given game"""
        pass
    
    def shuffle(self):
        random.shuffle(self.cards)
        self.free_cards = copy(self.cards)
    
    def __getitem__(self, position):
        return self.cards[position]

    def __len__(self):
        return len(self.cards)

    def __repr__(self):
        return f"{self.cards}"
    
    def take_cards(self, n):
        hand = random.sample(self.free_cards, n)
        for card in hand:
            self.free_cards.remove(card)
        return hand
    

class PokerDeck(BaseDeck):
    def build_deck(self):
        ranks = [f"{i}" for i in range(2,11)] + ["Jack", "Queen", "King", "Ace"]
        suits = ["Diamonds", "Clubs", "Hearts", "Spades"]
        return [Card(rank, suit) for rank in ranks for suit in suits]


class TH_Hand():
    def __init__(self, deck):
        self.cards = deck.take_cards(2)
    
    def __repr__(self):
        return f"{self.cards}"

class TH_Table():
    def __init__(self, deck: PokerDeck):
        self.cards = deck.take_cards(5)
        self.pointer = 0
    
    def __repr__(self):
        return f"{self.cards[:self.pointer]}"

    def show_cards(self, n):
        self.pointer += n
        return f"{self.cards[ : n]}"
    

class TH_game():
    def __init__(self, deck: PokerDeck , n_players: int):
        self.deck = deck
        self.deck.shuffle()
        self.hands = [TH_Hand(deck) for _ in range(n_players)]
        self.table = TH_Table(deck)
    
    def play(self):
        self.deck.shuffle()
        for i in range(len(self.hands)):
            print(f"Hand{i+1}: {self.hands[i]}")
        print(f"Table: {self.table.show_cards(5)}")

deck1 = PokerDeck()
deck1.build_deck()
game1 = TH_game(deck1, 3)

game1.play()
    
```
